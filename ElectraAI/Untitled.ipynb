```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import numpy as np

# 2.1 Hedefi tanımla (ikili)
TARGET = 'HighPerformance'
df[TARGET] = (df['PerformanceRating'] >= 4).astype(int)

# 2.2 K-Means'e girecek özellik setini hazırla
drop_cols = [TARGET, 'PerformanceRating', 'Attrition']  # hedef ve sızıntı olabilecek kolonlar
features_df = df.drop(columns=[c for c in drop_cols if c in df.columns], errors='ignore')

# 2.3 Tür ayrımı (drop sonrası!)
num_cols = features_df.select_dtypes(include=[np.number]).columns.tolist()
cat_cols = [c for c in features_df.columns if c not in num_cols]

# 2.4 Model ve kümeleme için iki farklı ön işleme pipeline'ı
# - supervised (ağaç tabanlı): sadece OHE (scale gerekmez)
pre_supervised = ColumnTransformer(
    transformers=[('cat', OneHotEncoder(handle_unknown='ignore', sparse_output=False), cat_cols)],
    remainder='passthrough'
)

# - unsupervised (k-means / lof): OHE + StandardScaler
pre_unsup = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore', sparse_output=False), cat_cols),
        ('num', StandardScaler(), num_cols)
    ],
    remainder='drop'
)

# 3.1 K-Means öncesi dönüşüm
X_unsup = pre_unsup.fit_transform(features_df)

# 3.2 k seçimi (elbow + silhouette)
sil_scores = {}
inertia = {}
for k in range(2, 9):
    km = KMeans(n_clusters=k, n_init=20, random_state=42)
    km.fit(X_unsup)
    inertia[k] = km.inertia_
    sil_scores[k] = silhouette_score(X_unsup, km.labels_)

print("Silhouette Scores:", sil_scores)
print("Inertia:", inertia)

# 3.3 En iyi k ile model ve etiketler
best_k = max(sil_scores, key=sil_scores.get)
kmeans = KMeans(n_clusters=best_k, n_init=50, random_state=42).fit(X_unsup)
df['cluster'] = kmeans.labels_

# 3.4 Küme profilleri (örnek özetler)
profile_cols = ['Department','JobRole','BusinessTravel','OverTime','EducationField']
print(df.groupby('cluster')[profile_cols].agg(lambda s: s.value_counts(normalize=True).head(3)))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA

# PCA ile 2 boyuta indirgeme
pca = PCA(n_components=2, random_state=42)
X2 = pca.fit_transform(X_unsup)

# Elbow & Silhouette
plt.figure(figsize=(12,4))
plt.subplot(121)
plt.plot(list(inertia.keys()), list(inertia.values()), marker='o')
plt.title('Elbow (Inertia)'); plt.xlabel('k'); plt.ylabel('Inertia')

plt.subplot(122)
plt.plot(list(sil_scores.keys()), list(sil_scores.values()), marker='o')
plt.title('Silhouette'); plt.xlabel('k'); plt.ylabel('Score')
plt.tight_layout(); plt.show()

# KMeans kümelerinin PCA alanında görselleştirilmesi
plt.figure(figsize=(6,5))
sns.scatterplot(x=X2[:,0], y=X2[:,1], hue=df['cluster'], s=30, palette="Set2")
plt.title('PCA – KMeans Kümeleri')
plt.xlabel('PC1'); plt.ylabel('PC2')
plt.legend(title='Cluster', bbox_to_anchor=(1.02, 1), loc='upper left')
plt.tight_layout(); plt.show()
```

```python
import pandas as pd
import numpy as np

# Sayısal değişkenlerde hedefle korelasyon
corrs = df.corr(numeric_only=True)[TARGET].sort_values(ascending=False)
print("Sayısal kolon korelasyonları:\n", corrs)

# 0.9 üzerindeki korelasyonları tespit et
high_corr_feats = corrs[abs(corrs) > 0.9].index.tolist()

# Hedefin kendisini çıkar
high_corr_feats = [f for f in high_corr_feats if f != TARGET]
print("\nHedefle yüksek korelasyonlu kolonlar (>0.9):", high_corr_feats)
```

```python
from sklearn.preprocessing import LabelEncoder

categorical_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()
cat_leak_feats = []

for col in categorical_cols:
    if col != TARGET:
        temp = df[[col, TARGET]].copy()
        if temp[col].nunique() < 50:  # sadece düşük kategori sayısında kontrol et
            le = LabelEncoder()
            temp[col] = le.fit_transform(temp[col])
            corr_val = np.corrcoef(temp[col], temp[TARGET])[0,1]
            if abs(corr_val) > 0.9:
                cat_leak_feats.append(col)

print("\nHedefle yüksek korelasyonlu kategorik kolonlar (>0.9):", cat_leak_feats)
```

```python
leakage_columns = list(set(high_corr_feats + cat_leak_feats))
print("\nÇıkarılacak sızıntı kolonları:", leakage_columns)

features = [c for c in df.columns if c not in leakage_columns + [TARGET, 'cluster']]
X = df[features]
y = df[TARGET]
```

```python
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_validate
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score, f1_score, confusion_matrix, classification_report

# Eğitim/test ayır
X_tr, X_te, y_tr, y_te = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

pipe_rf = Pipeline(steps=[
    ('prep', pre_supervised),  # senin preprocessing pipeline
    ('rf', RandomForestClassifier(
        n_estimators=400, max_depth=None, class_weight='balanced', random_state=42
    ))
])

# CV değerlendirme
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
scores = cross_validate(pipe_rf, X_tr, y_tr,
                        scoring=['roc_auc','f1','balanced_accuracy'],
                        cv=cv, n_jobs=-1, return_train_score=False)
print({m: scores[f'test_{m}'].mean().round(3) for m in ['roc_auc','f1','balanced_accuracy']})

# Fit + test
pipe_rf.fit(X_tr, y_tr)
probs = pipe_rf.predict_proba(X_te)[:,1]
preds = (probs >= 0.5).astype(int)

print("\nROC-AUC:", round(roc_auc_score(y_te, probs), 3))
print("F1:", round(f1_score(y_te, preds), 3))
print("Confusion Matrix:\n", confusion_matrix(y_te, preds))
print("\nClassification Report:\n", classification_report(y_te, preds))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_cluster_category_dist(df, col):
    """Verilen kategorik sütunun cluster bazında yüzdesel dağılımını çizer."""
    plt.figure(figsize=(8,5))
    
    # Cluster bazında yüzdelik dağılım
    cluster_dist = (
        df.groupby(['cluster', col])
        .size()
        .groupby(level=0)
        .apply(lambda x: x / x.sum())
        .to_frame(name='Percentage')  # Önce DataFrame'e çeviriyoruz
        .reset_index()                # Sonra index'i resetliyoruz
    )
    
    # Bar plot
    sns.barplot(data=cluster_dist, x='cluster', y='Percentage', hue=col)
    plt.title(f'{col} Dağılımı (Cluster Bazında)', fontsize=14, fontweight='bold')
    plt.ylabel('Yüzde')
    plt.xlabel('Cluster')
    plt.legend(title=col, bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()


# Örnek: Tüm kategorik kolonları otomatik çizdir
categorical_cols_to_plot = ['Department', 'JobRole', 'BusinessTravel', 'OverTime', 'EducationField']

for col in categorical_cols_to_plot:
    plot_cluster_category_dist(df, col)
```
